## iloc,  rows or slice or boolean mask or a lambda

## use  iloc[row][col] to extract a specific data element from a DataFrame

```python
rows,cols = d4.shape
for j in range(rows):
    rx = []
    for  i in range(cols):
        x = d4.iloc[j][i]
        if len(str(x)) == 3:
            x = " " + str(x)
        else:
            x = str(x)
        #x = str(j) + ',' +str(i)
        rx.append(x)
    print(rx)
    print('-------------------------------------')
```
